# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
oly = pd.read_csv('summer.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
print('Summer Olympics is held in ',
      len(oly['City'].unique()),
      'cities so far.')        # Summer Olympics is held in 22 cities so far
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = oly[oly['Medal']=='Gold']
dt = []
for i in oly['Sport'].unique():
    dt.append([i,len(df[df['Sport'] == i])])
dt = pd.DataFrame(dt,columns = ['Sport','Gold']).sort_values(by = 'Gold',ascending=False)
dt.head().plot(x = 'Sport',y = 'Gold',kind = 'bar',figsize = (5,5))
print(dt[dt['Gold']==dt['Gold'].max()]['Sport'].unique()[0], 'is having most number of Gold Medals so far.')
# Aquatics is having most number of Gold Medals so far.
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
dt = []
for i in oly['Sport'].unique():
    dt.append([i,len(oly[oly['Sport'] == i])])
dt = pd.DataFrame(dt,columns = ['Sport','Medal']).sort_values(by = 'Medal',ascending=False)
dt.head().plot(x = 'Sport',y = 'Medal',kind = 'bar',figsize = (5,5))
print(dt[dt['Medal']==dt['Medal'].max()]['Sport'].unique()[0], 'is having most number of Medals so far.')
```

### 4. Which player has won most number of medals? (Top 5)

```python
dt = []
for i in oly['Athlete'].unique():
    dt.append([i,len(oly[oly['Athlete'] == i])])
dt = pd.DataFrame(dt,columns = ['Athlete','Medal']).sort_values(by = 'Medal',ascending=False)
dt.head().plot(x = 'Athlete',y = 'Medal',kind = 'bar',figsize = (5,5))
print(dt[dt['Medal']==dt['Medal'].max()]['Athlete'].unique()[0], 'has won most number of Medals.')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df = oly[oly['Medal']=='Gold']
dt = []
for i in df['Athlete'].unique():
    dt.append([i,len(df[df['Athlete'] == i])])
dt = pd.DataFrame(dt,columns = ['Athlete','Gold']).sort_values(by = 'Gold',ascending=False)
dt.head().plot(x = 'Athlete',y = 'Gold',kind = 'bar',figsize = (5,5))
print(dt[dt['Gold']==dt['Gold'].max()]['Athlete'].unique()[0], 'has won most number of Golds.')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
dt = oly[oly['Medal'] == 'Gold']
print('In', dt[dt['Country']=='IND']['Year'].min(), 'India won first Gold Medal in Summer Olympics')
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
dt = []
for i in oly['Event'].unique():
    dt.append([i,len(oly[oly['Event'] == i]['Athlete'].unique())])
dt= pd.DataFrame(dt,columns =['Event','Athlete'])
dt.sort_values(by = 'Athlete',ascending=False).head().plot(x = 'Event',y = 'Athlete',kind = 'bar',figsize = (5,5))
print(dt[dt['Athlete'] == dt['Athlete'].max()]['Event'].unique()[0],'is most popular in terms on number of players')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df = oly[oly['Gender']=='Women']
df = df[df['Medal'] == 'Gold']
dt = []
for i in oly['Sport'].unique():
    dt.append([i,len(df[df['Sport'] == i])])
dt = pd.DataFrame(dt,columns = ['Sport','Gold']).sort_values(by = 'Gold',ascending=False)
dt.head().plot(x = 'Sport',y = 'Gold',kind = 'bar',figsize = (5,5))
print(dt[dt['Gold']==dt['Gold'].max()]['Sport'].unique()[0], 'is having most female Gold Medalists.')
```
